```python
# prompt: can you find out the mean median and mode for the salary column

# Calculate the median
median_salary = df['salary'].median()
print(f"Median Salary: {median_salary}")

# Calculate the mode
mode_salary = df['salary'].mode()
print(f"Mode Salary: {mode_salary}")
```

```python
# prompt: can you find out the basic stats information on my dataframe df , calculate  measurements such as  mean, median, mode, variance , skewness, range and kurtosis.

# Calculate the variance
variance_salary = df['salary'].var()
print(f"Variance of Salary: {variance_salary}")

# Calculate the skewness
skewness_salary = df['salary'].skew()
print(f"Skewness of Salary: {skewness_salary}")

# Calculate the range
range_salary = df['salary'].max() - df['salary'].min()
print(f"Range of Salary: {range_salary}")

# Calculate the kurtosis
kurtosis_salary = df['salary'].kurt()
print(f"Kurtosis of Salary: {kurtosis_salary}")
```

```python
# prompt: can you find out the basic stats information on my dataframe df, calculate measurements such as mean, median, mode, range, std, skewness and kurtosis

# Calculate the standard deviation
std_salary = df['salary'].std()


#You can extend this for other columns as needed e.g., 'service', 'phd' etc.
std_service = df['service'].std()

std_phd = df['phd'].std()
```

```python
# prompt: can  you find out eda on my dataframe df

import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical features
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.histplot(df['salary'], kde=True)
plt.title('Salary Distribution')

plt.subplot(1, 3, 2)
sns.histplot(df['service'], kde=True)
plt.title('Service Distribution')

plt.subplot(1, 3, 3)
sns.histplot(df['phd'], kde=True)
plt.title('PhD Distribution')
plt.tight_layout()
plt.show()

# Box plots for numerical features
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.boxplot(y=df['salary'])
plt.title('Salary Box Plot')

plt.subplot(1, 3, 2)
sns.boxplot(y=df['service'])
plt.title('Service Box Plot')

plt.subplot(1, 3, 3)
sns.boxplot(y=df['phd'])
plt.title('PhD Box Plot')
plt.tight_layout()
plt.show()

# Correlation matrix (heatmap)
plt.figure(figsize=(8, 6))
sns.heatmap(df[['salary', 'service', 'phd']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Pairplot for visualizing relationships between numerical variables
sns.pairplot(df[['salary', 'service', 'phd']])
plt.show()

# Explore categorical features (if any) using countplots or bar plots
# Example: Assuming 'rank' is a categorical column
# sns.countplot(x='rank', data=df)
# plt.title('Rank Distribution')
# plt.show()

#Further analysis based on specific questions or hypotheses
```
